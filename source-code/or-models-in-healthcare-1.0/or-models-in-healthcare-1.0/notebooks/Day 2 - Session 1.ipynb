## (3) What is NumPy and Why Use It?

NumPy (**Numerical Python**) is a powerful library for numerical computing in Python.

### üîπ Why NumPy?
- **Faster than lists**: Uses **C-based arrays** for speed.
- **Supports vectorized operations**: Perform operations on entire arrays without loops.
- **More functionality**: Includes linear algebra, random numbers, and efficient array operations.

We import NumPy as `np` for convenience (most people do it that way):

````python
import numpy as np

# Creating a simple NumPy array
arr = np.array([1, 2, 3, 4, 5])
print("NumPy Array:", arr)
````

#### üìå Example: Performance Comparison NumPy Arrays vs. Python Lists
NumPy arrays are faster and more memory-efficient than Python lists.

````python
import time

size = 1_000_000
python_list = list(range(size))
numpy_array = np.arange(size)

# Timing Python list multiplication
t0 = time.time()
python_result = [x * 2 for x in python_list]
t1 = time.time()
print("Python List Time:", t1 - t0, "seconds")

# Timing NumPy array multiplication
t0 = time.time()
numpy_result = numpy_array * 2
t1 = time.time()
print("NumPy Array Time:", t1 - t0, "seconds")
````

### üîπ Basic NumPy Operations: Creating Arrays
| Method | Description |
|--------|------------|
| `np.array([1,2,3])` | Convert list to array |
| `np.zeros((3,3))` | Create a 3x3 array of zeros |
| `np.ones((2,2))` | Create a 2x2 array of ones |
| `np.linspace(0, 10, 5)` | Create an array of 5 equally spaced numbers from 0 to 10 |
| `np.random.rand(3,3)` | Create a 3x3 array of random values |

#### üìå Example: Creating Different Arrays

````python
zeros = np.zeros((3, 3))
ones = np.ones((2, 2))
linspace = np.linspace(0, 10, 5)
random_array = np.random.rand(3, 3)

print("Zeros Array:\n", zeros)
print("Ones Array:\n", ones)
print("Linspace Array:", linspace)
print("Random Array:\n", random_array)
````

### üîπ NumPy Matrix Operations

| Operation | Description |
|-----------|------------|
| `A + B` | Matrix addition |
| `A - B` | Matrix subtraction |
| `A * B` | Element-wise multiplication |
| `A @ B` | Matrix multiplication |
| `A.T` | Transpose of matrix `A` |
| `np.linalg.inv(A)` | Inverse of matrix `A` |
| `np.linalg.det(A)` | Determinant of matrix `A` |

#### üìå Example: Matrix Operations

````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("A + B:\n", A + B)
print("A - B:\n", A - B)
print("Element-wise Multiplication:\n", A * B)
print("Matrix Multiplication (A @ B):\n", A @ B)
print("Transpose of A:\n", A.T)
print("Determinant of A:", np.linalg.det(A))
````

## (4) Introduction to Pandas DataFrames

Pandas is a powerful library for handling **structured data** using **DataFrames**.

### üîπ Why Pandas?
- **Easier data handling**: Works with tables, similar to Excel.
- **Fast and efficient**: Built on top of NumPy for speed.
- **Great for analysis**: Supports filtering, grouping, and aggregating data.

We import Pandas as `pd` for convenience:

````python
import pandas as pd

# Creating a simple DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 45],
    "Salary": [50000, 60000, 70000, 80000, 90000]
}
df = pd.DataFrame(data) # Create a DataFrame from a dictionary - keys become column names
                        # and values become column data
                        # Notice that df is now a DataFrame object
print("DataFrame:\n", df)
````

### üîπ Loading and Saving Data

Pandas can read and write data in various formats, such as CSV and Excel.

#### üìå Example: Reading and Writing Data

````python
# Writing DataFrame to Excel
df.to_csv("../data/sample_data.csv", index=False)  # Save without the index
                                                
print("DataFrame saved to './data/sample_data.csv'. Try to look in your data folder and see whether you can find the file.")

# Reading DataFrame from CSV
df_loaded = pd.read_csv("../data/sample_data.csv")
print("Loaded DataFrame:\n", df_loaded)
````

### üîπ Analyzing Data in Pandas

####  Basic DataFrame Operations
| Operation | Description |
|-----------|------------|
| `df.head()` | Show first 5 rows |
| `df.describe()` | Summary statistics |
| `df.info()` | Information about the DataFrame |
| `df.columns()` | A list of all the column names |
| `df.index()` | A list of all the indices |
| `df["column"].mean()` | Compute mean of a column |

#### üìå Example: Analyzing Data

1. Print out the head (the first 5 rows) of the DataFrame to Familiarize yourself with the output.

````python
print("First 5 rows:\n", df.head())
````

2. Print out summary statistics of the dataframe  - what do the individual rows mean?

````python
print("Summary statistics:\n", df.describe())
````

3. Determine the average of the column salary. 

````python
print("Average Salary:", df["Salary"].mean())
````

4. Compute salary as a fraction of total salary. 

````python
df["Salary_Fraction"] = df["Salary"] / df["Salary"].sum()
print("Updated DataFrame with Salary_Fraction:\n", df)
````

5. Append total row

````python
total_row = pd.DataFrame({
    "Name": ["Total"],
    "Age": ["-"],
    "Salary": [df["Salary"].sum()],
    "Salary_Fraction": [df["Salary_Fraction"].sum()]
})
df = pd.concat([df, total_row], ignore_index=True)
print("Updated DataFrame:\n", df)
````

## (5) Loading and Analyzing WHO Health Expenditure Data

The dataset we are working with comes from the **World Health Organization (WHO)** and contains **current health expenditure (CHE) in millions of USD per country and year since 2000**.

### üîπ Source
The data was retrieved from: [WHO Global Health Expenditure Database](https://apps.who.int/nha/database/Select/Indicators/en)

### üîπ What‚Äôs Inside the File?
- **Country**: Name of the country (Index)
- **Year**: The year of the data entry (Columns)
- **CHE (Mio USD)**: Current health expenditure in millions of USD (Values)

#### üìå Example: Load the dataset and explore its contents

````python

# Load the dataset

file_path = "../data/WHO_stats_CHE_in_MIO_USD.xlsx"
df_who = pd.read_excel(file_path)

# Display size of the dataset (rows, columns)
print("Size of the dataset:", df_who.shape)


# Display first few rows
print("First 5 rows of the dataset:\n", df_who.head())

#Display data of last 8 rows
#rint("First 8rows of the dataset:\n", df_who.tail(8))
````

### üîπ Filtering and Querying Data
We can filter the dataset to analyze specific years or countries.

#### üìå Example: Filtering Data
1.  Filter data for a specific country (e.g., Nepal)

````python
nepal_data = df_who[df_who["Countries"] == "Nepal"]
print("Health expenditure for the Nepal:\n", nepal_data)
#print("data shape",nepal_data.shape)
````

2. Retrieve health expenditure for Nepal in 2020 

````python
nepal_2020 =  df_who[df_who["Countries"] == "Nepal"]["2020"].values[0]  
print("Health expenditure for Nepal in 2020:", round(nepal_2020, 1), "Mio. USD.")
````

#### ‚úèÔ∏è Task: Explore the Dataset
1. Find the **total** health expenditure for all countries in **2020**. Round the result to 2 digits.


````python
# Total health expenditure for all countries in 2020
total_expenditure_2020 = df_who["2020"].sum()
print("Total Health Expenditure in 2020:", round(total_expenditure_2020, 2), "Mio. USD.")
````

2. Find the **country** with the highest health expenditure in **2020**.


````python
# Country with the highest health expenditure in 2020
expenditure_2020 = df_who["2020"] # Extract the 2020 column
highest_country_2020 = expenditure_2020.idxmax() # Get the index of the maximum value
highest_country_2020 = df_who["Countries"][highest_country_2020] # Get the country name
print("\nCountry with highest expenditure in 2020:", highest_country_2020)
````

3. Compute the **average** health expenditure across all years for a selected country (e.g., Germany).


````python
# Average expenditure across all years for Germany
germany_data = df_who[df_who["Countries"] == "Germany"].iloc[:, 1:] # Extract data for Germany
avg_expenditure_germany = germany_data.mean(axis=1).values[0] # Calculate the average, axis=1 means row-wise
print("\nAverage health expenditure in Germany:", avg_expenditure_germany)
````

4. Find the **top 5 countries** with the highest average health expenditure.

````python
df_who["Avg_Expenditure"] = df_who.iloc[:, 1:].mean(axis=1) # Calculate average expenditure across all years in an extra column
top_countries = df_who.nlargest(5, "Avg_Expenditure")[["Countries", "Avg_Expenditure"]] 
print("\nTop 5 Countries with Highest Average Health Expenditure:\n", top_countries)
````

## (5) Introduction to Matplotlib & Seaborn

Matplotlib and Seaborn are two of the most commonly used Python libraries for **data visualization**.

### üîπ Matplotlib
- A flexible library for creating static, animated, and interactive visualizations.
- Works with NumPy and Pandas for seamless integration.

### üîπ Seaborn
- Built on top of Matplotlib, Seaborn makes statistical visualizations **simpler and more attractive**.
- Provides additional features such as built-in themes and better handling of categorical data.

#### üìå Example: Importing Matplotlib and Seaborn

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set_style("whitegrid")
````

### üîπ Creating a Bar Plot

We will create a **bar plot** showing the health expenditures of different countries in 2020, sorted from highest to lowest.

#### üìå Example: Bar Plot of Health Expenditure

````python
# Sort the DataFrame by the latest available year
df_2020 = df_who[["Countries", "2020"]] # Extract the 2020 column
# Sort the DataFrame by 2020
sorted_df = df_2020.sort_values(by="2020", ascending=False)

# Plot a bar chart
plt.figure(figsize=(20, 6)) # Set the figure size for the plot
sns.barplot(data = sorted_df, y="2020", x="Countries")

# Customize the plot
plt.xlabel("Country") # Set the label for the x-axis
plt.ylabel("Health Expenditure (Mio USD)")
plt.title("Health Expenditure by Country in 2020")
plt.show()
````

Let's face it! The above is not a very nice plot. There are too many countries on the x-axis, the labels on the x-axis overlap and the US is dominating the plot. 

Let's fix this step by step.

1. Restrict the plot to the top 30 countries, so that there is sufficient space. 

````python
# reduce to first 30 countries
print(sorted_df.shape)
sorted_df = sorted_df[:30]
print(sorted_df.shape)

# Plot a bar chart
plt.figure(figsize=(20, 6)) # Set the figure size for the plot
sns.barplot(data = sorted_df, y="2020", x="Countries")

# Customize the plot
plt.xlabel("Country") # Set the label for the x-axis
plt.ylabel("Health Expenditure (Mio USD)")
plt.title("Health Expenditure by Country in 2020")
plt.show()
````

2. Now tilt the x-lables by 80 degrees so that they do no longer overlap.

````python
# Plot a bar chart
plt.figure(figsize=(20, 6)) # Set the figure size for the plot
sns.barplot(data = sorted_df, y="2020", x="Countries")

# Customize the plot
plt.xlabel("Country") # Set the label for the x-axis
plt.xticks(rotation = 80) # Rotate the x-axis ticks by 80 degrees
plt.ylabel("Health Expenditure (Mio USD)")
plt.title("Health Expenditure by Country in 2020")
plt.show()
````

3. Change the y-axis to a logarithmic scale. 

````python
# Plot a bar chart
plt.figure(figsize=(20, 6)) # Set the figure size for the plot
sns.barplot(data = sorted_df, y="2020", x="Countries")

# Customize the plot
plt.xlabel("Country") # Set the label for the x-axis
plt.xticks(rotation = 80) # Rotate the x-axis ticks by 80 degrees
plt.ylabel("Health Expenditure (Mio USD)")
plt.yscale("log") # Set the y-axis to logarithmic scale
plt.title("Health Expenditure by Country in 2020")
plt.show()
````

### üîπ Creating a Line Plot: Health Expenditure Trend in Nepal

We will now create a **line plot** showing the trend of **Nepal‚Äôs health expenditure** over the years.

#### üìå Example: Line Plot of Nepal‚Äôs Health Expenditure

````python
# Extract Nepal‚Äôs data
nepal_data = df_who[df_who["Countries"] == "Nepal"].iloc[:, 1:].T  # Transpose to get years on x-axis
nepal_data.columns = ["Health Expenditure"]

# Plot the line chart
plt.figure(figsize=(10, 5))
sns.lineplot(data=nepal_data, marker="o")

# Customize the plot
plt.xlabel("Year")
plt.ylabel("Health Expenditure (Mio USD)")
plt.title("Health Expenditure Trend in Nepal (2000-Present)")
plt.xticks(rotation=45)
plt.show()
````
